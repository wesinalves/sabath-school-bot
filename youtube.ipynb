```python
def get_video(channel, publish_time, youtube_api):
    url = "https://youtu.be/"
    request = youtube.search().list(
        part="snippet",
        maxResults=25,
        q="escola sabatina",
        type="video",
        channelId=channel,
        publishedAfter=publish_time
    )
    response = request.execute()
    title = response['items'][0]['snippet']['title']
    url = url + response['items'][0]['id']['videoId']
    
    return (title, url)
    #return response
```
